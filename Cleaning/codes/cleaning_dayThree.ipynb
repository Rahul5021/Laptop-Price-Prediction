```python
import re
all_laptop_series = [
    'Spectre', 'Envy', 'Pavilion', 'Omen',  # HP
    'ThinkPad', 'IdeaPad', 'Legion',  # Lenovo
    'Aspire', 'Predator', 'Swift',  # Acer
    'MacBook Air', 'MacBook Pro',  # Apple
    'XPS', 'Inspiron', 'Alienware',  # Dell
    'GS', 'GE', 'Creator',  # MSI
    'ZenBook', 'ROG', 'VivoBook',  # ASUS
    'Realme Book',  # Realme
    'InBook X1',  # Infinix
    'Galaxy Book',  # Samsung
    'Primebook',  # Primebook
    'Herobook',  # CHUWI
    'AXL',  # AXL
    'AERO', 'AORUS',  # GIGABYTE
    'Ultimus',  # Ultimus
    'Alienware'  # ALIENWARE
]
no_type = 0
type = 0
list_laptop_type = []
# Extract laptop type for each product name

for name in df["Product Name"]:
    for laptop_type in all_laptop_series:
        if re.search(rf'\b{re.escape(laptop_type.lower())}\b', name.lower()):
            list_laptop_type.append(laptop_type)
            type += 1
            print(f"Matched: {name} -> {laptop_type}")
            break
    else:
        list_laptop_type.append("N\A")
        no_type += 1
        print(f"No match found for: {name}")
```

```python
df["Company"].unique()
```

```python
print(list_laptop_type)
```

```python
unique_company = set()
for company in list_laptop_type:
    unique_company.add(company)
print(unique_company)
```

```python
len(list_laptop_type)
```

```python
gaming_laptops = ["Omen", "Legion", "Predator", "Alienware", "GS", "GE", "ROG", "AORUS"]
for index, laptop_type in enumerate(list_laptop_type):
    for gaming in gaming_laptops:
        if gaming.lower() == laptop_type.lower():
            list_laptop_type[index] = "Gaming"
```

```python
df["Type"] = list_laptop_type
```
