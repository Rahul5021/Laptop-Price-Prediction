```python
import pandas as pd
```

```python
df = pd.read_csv("dayOne_cleaning.csv")
```

```python
import ast
df["Features"] = df["Features"].apply(ast.literal_eval)
```

```python
for feature in df["Features"]:
    print(type(feature))
```

```python
os_feature_list = []

for feature in df["Features"]:
    found_os = False  # Flag to track if "Processor" is found in the current feature list
    for individual_feature in feature:
        if "Operating System" in individual_feature:
            if not found_os:
                os_feature_list.append(individual_feature)
                found_os = True
```

```python
len(os_feature_list)
```

```python
df["Operating System"] = os_feature_list
```

```python
df.head()
```

```python
ram_feature_list = []

for feature in df["Features"]:
    found_ram = False  # Flag to track if "Processor" is found in the current feature list
    for individual_feature in feature:
        if "ram" in individual_feature.lower():
            if not found_ram:
                ram_feature_list.append(individual_feature)
                found_ram = True
```

```python
len(ram_feature_list)
```

```python
df["RAM"] = ram_feature_list
```

```python
df.head()
```

```python
for ram in df["RAM"]:
    print(type(ram))
```

```python
df.head()
```

```python
display_feature_list = []

for feature in df["Features"]:
    found_display = False  # Flag to track if "Processor" is found in the current feature list
    for individual_feature in feature:
        if "Display" in individual_feature:
            if not found_display:
                display_feature_list.append(individual_feature)
                found_display = True
```

```python
len(display_feature_list)
```

```python
df["Display"] = display_feature_list
```

```python
df.head()
```

```python
storage_feature_list = []

for feature_list in df["Features"]:
    storage_flag = False
    for individual_feature in feature_list:
        if "SSD" in individual_feature or "HDD" in individual_feature:
            storage_feature_list.append(individual_feature)
            storage_flag = True
            break
    if not storage_flag:
        storage_feature_list.append("N/A")
```

```python
storage_feature_list = []

for feature_list in df["Features"]:
    storage_details = [individual_feature for individual_feature in feature_list if "SSD" in individual_feature or "HDD" in individual_feature]

    # Check if storage details were found
    if storage_details:
        storage_feature_list.append(storage_details)
    else:
        # Append "N/A" if no storage details were found
        storage_feature_list.append("N/A")

# Print the results for each row
for i, features in enumerate(df["Features"]):
    print(f"Row {i + 1}: {features} => Storage Details: {storage_feature_list[i]}")
```

```python
df["Storage"] = storage_feature_list
```
