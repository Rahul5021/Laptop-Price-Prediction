```python

import re
processor_types = []
for processor in df["Processor"]:
    match = re.search(r'(Ryzen \d+|i\d+|Athlon Dual Core|M1)', processor)
    if match:
        processor_types.append(match.group())
    else:
        processor_types.append('Other')

print(processor_types)
```

```python
df["Processor Type"] = processor_types
```
