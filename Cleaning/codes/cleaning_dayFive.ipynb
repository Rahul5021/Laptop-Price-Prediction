```python
import re

def extract_inches(text):
    match = re.search(r'(\d+(?:\.\d+)?)\s*(?:Inch|"|\'|Inch)', text)
    return match.group(1) if match else None
```

```python
df["Inches"] = df["Display"].apply(extract_inches)
```

```python
pattern = re.compile(r'(\d+)\s?TB\s?HDD', re.IGNORECASE)
```

```python
hdd_storage = [pattern.search(s).group(1) if pattern.search(s) else "N\A" for s in hdd_storage]
```

```python
pattern = re.compile(r'(\d+)\s?(TB|GB)\s?SSD', re.IGNORECASE)
```

```python
import re
```

```python
ssd_storage = [(match.group(1), match.group(2)) if match else ("N\A", "N\A") for ssd in df["Storage"] for match in [pattern.search(ssd)]]
```

```python
for ssd in ssd_storage:
    print(ssd)
```

```python
formatted_ssd_storage = [f"{size} {unit}" for size, unit in ssd_storage]
```

```python
formatted_ssd_storage
```

```python
for index, ssd in enumerate(formatted_ssd_storage):
    unit = ssd.split(" ")[1]
    if unit.lower() == "tb":
        size = int(ssd.split(" ")[0]) * 1024
        formatted_ssd_storage[index] = f"{size} GB"
```

```python
formatted_ssd_storage
```

```python
for index, na in enumerate(formatted_ssd_storage):
    if "N\A N\A" in na:
        formatted_ssd_storage[index] = "N\A"
```

```python
capacities_in_gb = []
for capacity in formatted_ssd_storage:
    if capacity == "N\A":
        capacities_in_gb.append(capacity)
    else:
        capacities_in_gb.append(int(capacity.split(" ")[0]))
```

```python
capacities_in_gb
```

```python
df["SSD"] = capacities_in_gb
```
